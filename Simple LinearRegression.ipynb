```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
Shirisha=pd.read_csv(r"C:\Users\nirma\Desktop\SHIIRISHA\Simple Linear Regression.csv")
```

```python
Shirisha
```

```python
lr=LinearRegression()
```

```python
x=Shirisha.iloc[:,0].values.reshape(-1,1)
y=Shirisha.iloc[:,1].values.reshape(-1,1)
```

```python
lr.fit(x,y)
```

```python
y_predict=lr.predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x,y_predict,color='y')
plt.xlabel("GATE")
plt.ylabel("GPA")
plt.title("GATE, GPA Prediction Model")
```
